```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("US_Accidents_Small.csv")
df.head()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour

plt.figure(figsize=(10,5))
sns.histplot(df['Hour'], bins=24)
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Weather_Condition", data=df)
plt.xticks(rotation=45)
plt.title("Accidents by Weather Condition")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Weather_Condition", hue="Severity", data=df)
plt.xticks(rotation=45)
plt.title("Severity vs Weather")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Sunrise_Sunset", data=df)
plt.title("Day vs Night Accidents")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Visibility(mi)'], bins=20)
plt.title("Visibility Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Precipitation(in)'], bins=20)
plt.title("Precipitation Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    x=df['Start_Lng'],
    y=df['Start_Lat'],
    cmap="Reds",
    fill=True
)
plt.title("Accident Hotspots")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['State'].value_counts().plot(kind='bar')
plt.title("Accidents by State")
plt.show()
```
